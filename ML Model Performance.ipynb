```python
import pandas as pd
import numpy as np
```

```python
dfH =  pd.read_csv('Hour Price Prediction -lessAtt.csv')
```

```python
dfH.shape
```

```python
dfH
```

# Pre processing

```python
dfH.drop(["Id"], axis=1, inplace=True) #deleting unnecessary coloumn
dfH.drop(["3SsnPorch"], axis=1, inplace=True)
dfH.drop(["PoolArea"], axis=1, inplace=True)
dfH.drop(["ScreenPorch"], axis=1, inplace=True)
```

```python
x = dfH.drop(['SalePrice'],axis=1)
y = dfH['SalePrice']
```

# Linear regression with hold out menthod for CV

```python
from sklearn.model_selection import train_test_split
Tx, Tex, Ty,Tey = train_test_split(x,y, test_size=0.3, random_state=100) #model selection
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression() 
modelLR.fit(Tx,Ty)
predicLR = modelLR.predict(Tex)
from sklearn import metrics
print("Simple Linear Regression")
print("MAE:", metrics.mean_absolute_error(predicLR,Tey))
print("MSE:", metrics.mean_squared_error(predicLR,Tey))
print("RMSE:", np.sqrt(metrics.mean_squared_error(predicLR,Tey)))
print("Coeff of determination:", metrics.r2_score(predicLR,Tey))
```

# Decision Tree with hold out method for CV

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier(criterion='entropy',max_depth=10)
modelDT.fit(Tx,Ty)
precDT = modelDT.predict(Tex)
print("Decision Tree Classifier")
print("MAE:", metrics.mean_absolute_error(precDT,Tey))
print("MSE:", metrics.mean_squared_error(precDT,Tey))
print("RMSE:", np.sqrt(metrics.mean_squared_error(precDT,Tey)))
print("Coeff of determination:", metrics.r2_score(precDT,Tey))
```

# Random Forest without cross validation with hold out method for CV

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(Tx,Ty)
predictRF = model.predict(Tex)
print("Random Forest")
print("MAE:", metrics.mean_absolute_error(predictRF,Tey))
print("MSE:", metrics.mean_squared_error(predictRF,Tey))
print("RMSE:", np.sqrt(metrics.mean_squared_error(predictRF,Tey)))
print("Coeff of determination:", metrics.r2_score(predictRF,Tey))
```
